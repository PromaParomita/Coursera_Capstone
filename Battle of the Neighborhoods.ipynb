```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) 


filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['address', 
                  'address latitude', 
                  'address longitude', 
                  'venue', 
                  'venue latitude', 
                  'venue longitude', 
                  'venue category']
    
    return(nearby_venues)
```

```python
charlesmark_venues = getNearbyVenues(names=Boston_Hotel['address'],
                                   latitudes=Boston_Hotel['latitude'],
                                   longitudes=Boston_Hotel['longitude']
                                  )
```

```python
charlesmark_onehot = pd.get_dummies(charlesmark_venues[['venue category']], prefix="", prefix_sep="")


charlesmark_onehot['address'] = charlesmark_venues['address'] 

fixed_columns = [charlesmark_onehot.columns[-1]] + list(charlesmark_onehot.columns[:-1])
charlesmark_onehot = charlesmark_onehot[fixed_columns]

charlesmark_onehot.head()
```

```python
charlesmark_grouped = charlesmark_onehot.groupby('address').mean().reset_index()
charlesmark_grouped
```

```python

num_top_venues = 5

for hood in charlesmark_grouped['address']:
    print("----"+hood+"----")
    temp = charlesmark_grouped[charlesmark_grouped['address'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

creating top 10 values

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']


columns = ['address']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))


address_venues_sorted = pd.DataFrame(columns=columns)
address_venues_sorted['address'] = charlesmark_grouped['address']

for ind in np.arange(charlesmark_grouped.shape[0]):
    address_venues_sorted.iloc[ind, 1:] = return_most_common_venues(charlesmark_grouped.iloc[ind, :], num_top_venues)

address_venues_sorted.head()
```

```python
kclusters = 5

charlesmark_grouped_clustering = charlesmark_grouped.drop('address', 1)


kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(charlesmark_grouped_clustering)


kmeans.labels_[0:10] 
```

```python
 
address_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
Boston_merged = Boston_Hotel


Boston_merged = Boston_merged.join(address_venues_sorted.set_index('address'), on='address')

Boston_merged.head() 
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]


markers_colors = []
for lat, lon, poi, cluster in zip(Boston_merged['latitude'], Boston_merged['longitude'], Boston_merged['address'], Boston_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
